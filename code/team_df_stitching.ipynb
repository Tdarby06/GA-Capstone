```python
col_order_string = ['game#', 'day', 'date', 'team', 'home_away', 'opp', 'win_loss', 'r', 
        'ra', 'innings', 'record', 'divison_rank', 'games_back', 'winning_pitcher', 
        'losing_pitcher', 'game_duration', 'day_night', 'attendance', 'streak', 'saving_pitcher', 
        'bat_hits', 'bat_rbi', 'bat_walks', 'bat_strike_outs', 'hou_plate_app', 'bat_batting_avg', 
        'bat_obp', 'bat_slg', 'bat_ops', 'pitch_opp_hits', 'pitch_walks', 'pitch_strike_outs', 
        'pitch_opp_hr', 'pitch_era', 'pitch_batters_faced', 'pitch_total_pitches', 'pitch_combined_strikes', 
        'pitch_strikes_contacted', 'pitch_strikes_swinging', 'pitch_strikes_looking', 'pitch_wpa']
```

```python
def year_check(club):
    year_list = [1995, 1996, 1997, 1998, 1999,
             2000, 2001, 2002, 2003, 2004,
             2005, 2006, 2007, 2008, 2009,
             2010, 2011, 2012, 2013, 2014,
             2015, 2016, 2017, 2018, 2019]
    missing_years = []
    club['date'] = pd.to_datetime(club['date'])
    for n in year_list:
        if club[club['date'].dt.year == n].shape[0] == 0:
            missing_years.append(n)
        else:
            pass
    first_season = min(club['date'])
    last_season = max(club['date'])
    return missing_years, first_season, last_season
```

```python
# single year merge function
def single_year_merge(df1, df2):
    boxscore_cols = ['date', 'bat_hits', 'bat_rbi', 'bat_walks',
       'bat_strike_outs', 'hou_plate_app', 'bat_batting_avg', 'bat_obp',
       'bat_slg', 'bat_ops', 'pitch_opp_hits', 'pitch_walks',
       'pitch_strike_outs', 'pitch_opp_hr', 'pitch_era', 'pitch_batters_faced',
       'pitch_total_pitches', 'pitch_combined_strikes',
       'pitch_strikes_contacted', 'pitch_strikes_swinging',
       'pitch_strikes_looking', 'pitch_wpa']
    #change boxscore column from day to date
    df2.rename(columns= {'day' : 'date'}, inplace=True)
    #put df1 and df2 dates into datetime
    df1['date'] = pd.to_datetime(df1['date'])
    df2['date'] = pd.to_datetime(df2['date'])
    #merge df1 and df2 
    df = pd.merge(df1.drop(labels = 'boxscore', axis = 1), df2[boxscore_cols], how = 'outer', on = 'date')
    df.drop_duplicates(subset=['date'], inplace = True)
    #asking for user input for file name
    print('give team name and year for file name:')
    a = input()
    b = input()
    df.to_csv(f'../data/pre_clean_data/{a.upper()}/{a}_{b}.csv', index=False)
    return print('saved!')
```
